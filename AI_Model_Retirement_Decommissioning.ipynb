```python
import pandas as pd
import datetime
import random

print("AI Retirement Governance System Initialized")
```

```python
models = pd.DataFrame([
    ["Fraud Detection v1", "Production", 0.92, "Active"],
    ["Recommendation v2", "Production", 0.71, "Active"],
    ["Chatbot Legacy", "Production", 0.55, "Active"]
], columns=["Model", "Stage", "Performance", "Status"])

models
```

```python
PERFORMANCE_THRESHOLD = 0.75

def detect_retirement_triggers(df):
    df["Trigger"] = df["Performance"].apply(
        lambda x: "Performance Degradation" if x < PERFORMANCE_THRESHOLD else "Healthy"
    )
    return df

models = detect_retirement_triggers(models)
models
```

```python
decommission_log = []

def decommission_model(model_name):
    steps = [
        "API Access Revoked",
        "Traffic Redirected",
        "Dependencies Removed",
        "Rollback Disabled",
        "Model Permanently Disabled"
    ]

    for step in steps:
        decommission_log.append({
            "Model": model_name,
            "Step": step,
            "Timestamp": datetime.datetime.now()
        })

    models.loc[models["Model"] == model_name, "Status"] = "Retired"

# Retire triggered models
for m in models[models["Trigger"]=="Performance Degradation"]["Model"]:
    decommission_model(m)

pd.DataFrame(decommission_log)
```

```python
archive = {}

def archive_model(model):
    archive[model] = {
        "model_artifacts": f"{model}_weights.pkl",
        "training_data": f"{model}_dataset.csv",
        "evaluation_report": f"{model}_metrics.pdf",
        "retention_years": 7
    }

for m in models[models["Status"]=="Retired"]["Model"]:
    archive_model(m)

archive
```
