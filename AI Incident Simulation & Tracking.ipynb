```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

# Create Incident Event Log

```python
incident_events = [
    {"time": "10:00", "event": "Bias monitoring alert triggered", "severity": "SEV-2"},
    {"time": "10:07", "event": "Incident Commander assigned", "severity": "SEV-2"},
    {"time": "10:18", "event": "Containment: Model rollback initiated", "severity": "SEV-2"},
    {"time": "11:05", "event": "Impact wider than expected", "severity": "SEV-1"},
    {"time": "11:40", "event": "Investigation: Root cause analysis started", "severity": "SEV-1"},
    {"time": "14:10", "event": "Mitigation: Guardrail patch deployed", "severity": "SEV-2"},
    {"time": "16:00", "event": "System stabilized", "severity": "SEV-3"},
]

df = pd.DataFrame(incident_events)
df
```

# Convert Time → Timestamp

```python
base_date = datetime.today().date()

df["timestamp"] = df["time"].apply(
    lambda t: datetime.strptime(f"{base_date} {t}", "%Y-%m-%d %H:%M")
)

df
```

# Calculate Minutes from Incident Start

```python
df["minutes_from_start"] = (
    df["timestamp"] - df["timestamp"].iloc[0]
).dt.total_seconds() / 60

df
```

# Severity Escalation Summary

```python
severity_counts = df["severity"].value_counts()
severity_counts
```

# Key Effectiveness Metrics

```python
containment_time = df[df["event"].str.contains("Containment")]["minutes_from_start"].values[0]
investigation_time = df[df["event"].str.contains("Investigation")]["minutes_from_start"].values[0]
mitigation_time = df[df["event"].str.contains("Mitigation")]["minutes_from_start"].values[0]

print("Time to Containment:", containment_time, "minutes")
print("Time to Investigation:", investigation_time, "minutes")
print("Time to Mitigation:", mitigation_time, "minutes")
```
